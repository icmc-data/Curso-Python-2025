````python
anscombe = sns.load_dataset('anscombe')

# pontos de I at√© IV
X1 = anscombe[anscombe.dataset == 'I'].x.values
Y1 = anscombe[anscombe.dataset == 'I'].y.values

X2 = anscombe[anscombe.dataset == 'II'].x.values
Y2 = anscombe[anscombe.dataset == 'II'].y.values

X3 = anscombe[anscombe.dataset == 'III'].x.values
Y3 = anscombe[anscombe.dataset == 'III'].y.values

X4 = anscombe[anscombe.dataset == 'IV'].x.values
Y4 = anscombe[anscombe.dataset == 'IV'].y.values

# Linha vermelha
line_x = np.linspace(3, 20, 100)
line_y = 3 + 0.5 * line_x

# Plotar!
````

## Gabarito:

````python
fig, axs = plt.subplots(2, 2)

fig.suptitle('Anscombe')

axs[0, 0].set_title('I')
axs[0, 0].scatter(X1, Y1)
axs[0, 0].plot(line_x, line_y, c='red')

axs[0, 1].set_title('II')
axs[0, 1].scatter(X2, Y2)
axs[0, 1].plot(line_x, line_y, c='red')

axs[1, 0].set_title('III')
axs[1, 0].scatter(X3, Y3)
axs[1, 0].plot(line_x, line_y, c='red')

axs[1, 1].set_title('IV')
axs[1, 1].scatter(X4, Y4)
axs[1, 1].plot(line_x, line_y, c='red')
````

# Recrie essa imagem:
Sem t√≠tulo.png

````python
titanic = sns.load_dataset('titanic')

idades = titanic.age.values
generos = ['Mulheres', 'Homens']
quant_generos = titanic.sex.value_counts().values

# Plotar!
````

## Gabarito

````python
fig, axs = plt.subplots(1,2)
fig.set_size_inches(8, 4)

axs[0].hist(idades, bins=10)
axs[0].set_title('Idades de pessoas no Titanic')
axs[0].set_xlabel('Idade')
axs[0].set_ylabel('Frequ√™ncia')

axs[1].bar(generos, quant_generos, color=['red', 'blue'])
axs[1].set_title('G√™nero de pessoas no Titanic')
axs[1].set_ylabel('Frequ√™ncia')
````

# Recrie essa imagem

titanic_por_idade.png

````python
homens_menos_30 = titanic[(titanic['age'] <= 30) & (titanic['sex'] == 'male')].shape[0]
homens_mais_30 = titanic[(titanic['age'] > 30) & (titanic['sex'] == 'male')].shape[0]

mulheres_menos_30 = titanic[(titanic['age'] <= 30) & (titanic['sex'] == 'female')].shape[0]
mulheres_mais_30 = titanic[(titanic['age'] > 30) & (titanic['sex'] == 'female')].shape[0]

generos = ['Mulheres', 'Homens']
genero_30_ou_menos = [mulheres_menos_30, homens_menos_30]
genero_mais_que_30 = [mulheres_mais_30, homens_mais_30]
````

## Gabarito

````python
x = np.arange(len(generos))
largura = 0.35


fig, ax = plt.subplots()
ax.bar(x - largura/2, genero_30_ou_menos, largura, label='30 anos ou menos', color='blue')
ax.bar(x + largura/2, genero_mais_que_30, largura, label='Mais de 30 anos', color='orange')


ax.set_title('Distribui√ß√£o de Homens e Mulheres por Faixa Et√°ria no Titanic')
ax.set_xticks(x)
ax.set_xticklabels(generos)
ax.set_xlabel('G√™nero')
ax.set_ylabel('N√∫mero de Pessoas')


ax.legend()


plt.show()
````
